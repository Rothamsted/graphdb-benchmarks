```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


d = { 
  "shop": [ "London", "Berlin", "Paris" ],
  "sales-2020": [ 1000, 2344, 1233 ],
  "sales-2021": [ 355, 4003, 2344 ],
  "sales-2022": [ 2344, 2949, 3443 ]
}

colors = { "2020": "Blue", "2021": "Green", "2022": "Red" }

df = pd.DataFrame ( d, index = d [ "shop" ] )

ax = None
for yk in [ 2020, 2021, 2022 ]:
  yk = str ( yk )
  ax = df.plot.bar ( y = "sales-" + yk, label = yk, color = colors [ yk ], stacked = False, ax = ax )
    
    
```
